# **Input Data**

```python
#Load Python Package
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Input data terdiri  dari beberapa bagian, yaitu menjalankan *package* yang digunakan, memasukkan dataset yang akan digunakan dan mengatur data yang akan digunakan. Pada kasus ini, data yang digunakan merupakan hasil pada kolom ke-3 dan ke-4 dari dataset Pengunjung Mall yang telah diunggah.

# **Metode K-Means Clustering**

```python
#Load Phyton Package
from sklearn.cluster import KMeans
```

```python
#Optimisasi K-Means dengan Metode Elbow
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
#Visualisasi Hasil Optimisasi
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Cluster Number')
plt.ylabel('WCSS')
plt.vlines(5, min(wcss), max(wcss), color='red', linestyle = 'dotted')
plt.vlines(3, min(wcss), max(wcss), color='red', linestyle = 'dotted')
plt.show()
```

```python
#K-Means Clustering (5 Cluster)
kmeans = KMeans(n_clusters = 5, init = 'k-means++',
random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
#Visualisasi Hasil Cluster
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'blue', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'red', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'magenta', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'green', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Consumers Cluster')
plt.xlabel('Yearly Salary')
plt.ylabel('Yearly expense rating (1-100)')
plt.legend()
```

```python
#K-Means Clustering (3 Cluster)
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
#Visualisasi Hasil Cluster
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'blue', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'red', label = 'Custer 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'magenta', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Consumers Cluster')
plt.xlabel('Yearly Salary')
plt.ylabel('Yearly expense rating (1-100)')
plt.legend()
```

Metode K-Means Clustering terdiri dari beberapa bagian, yaitu menjalankan *package*, mencari jumlah *cluster* yang terbaik dengan menggunakan metode Elbow, dan membuat *cluster* berdasarkan hasil yang diperoleh dari metode Elbow. Visualisai hasil optimisasi menunjukkan *cluster* yang baik adalah K=3 dan K=5 karena reduksi variasi cukup besar. Dengan demikian, data dikelompokkan menjadi 3 dan 5 *cluster*.

# **Metode Hierarchical Clustering**

```python
#Load Python Package
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
#Visualisasi Dendogram
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Consumer')
plt.ylabel('Euclidean Distance')
plt.hlines(230, 0, 10000, color='red', linestyle = 'dotted')
plt.show()
```

```python
#Agglomerative Hierarchical Clustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
y_hc
```

```python
#Visualisasi Hasil Cluster
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Consumers Cluster')
plt.xlabel('Yearly Salary')
plt.ylabel('Yearly expense rating (1-100)')
plt.legend()
plt.show()
```

Metode Hierarchical Clustering terdiri dari beberapa bagian, yaitu menjalankan *package*, visualisasi dendogram untuk menentukan jumlah cluster yang terbaik, membuat Hierarchical Clustering dengan Agglomerative (*bottom-up)* berdasarkan jumlah cluster yang terbaik. Menentukan jumlah cluster pada visualisasi dendogram menggunakan caara membuat garis horizontal pada jarak 230. Garis tersebut menunjukkan jumlah cluster sebanyak 3 cluster. Dengan demikian,  Hierarchical Clustering dengan Agglomerative dibuat sebanyak 3 cluster.

# **Metode DBSCAN**

```python
#Load Python Package
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
#Generate Data
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(n_samples = 750, centers = centers, cluster_std = 0.4, random_state = 0)
X = StandardScaler().fit_transform(X)
```

```python
#Metode DBSCAN
db = DBSCAN(eps = 0.3, min_samples = 10).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
```

```python
#Penentuan Cluster, Pencilan, dan Homogenitas
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
```

```python
#Visualisasi Hasil Cluster
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
  if k == -1:
    col = [0, 0, 0, 1]
  class_member_mask = (labels == k)
  xy = X[class_member_mask & core_samples_mask]
  plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=14)
  xy = X[class_member_mask & ~core_samples_mask]
  plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)
plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

Metode DBSCAN terdiri dari beberapa bagian, yaitu menjalankan *package*, membuat data set secara acak, menjalankan metode DBSCAN, menentukan cluster, pencilan, dan homogenitas, dan membuat visualisasi. Metode DBSCAN menerapkan radius 0.3 dan minimum yang dikelompokkan 10 data.


# **Latihan Penerapan Metode Clustering**

```python
#Generate Data
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
X = StandardScaler().fit_transform(X)
```

```python
#Metode DBSCAN
db = DBSCAN(eps = 0.3, min_samples = 11).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
#Penentuan Cluster, Pencilan, dan Homogenitas
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
print("Homogeneity: %0.3f" % metrics.homogeneity_score(y, labels))
```

```python
# Visualisasi hasil clustering
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.show()
```

Latihan penerapan metode clustering agar sesuai dengan hasil yang tertera menerapkan metode DBSCAN. Metode DBSCAN pada kasus ini menerapkan radius 0.3 dan minimum yang dikelompokkan 11 data. Namun, hasil yang diperoleh tidak sama persis dengan hasil yang tertera disebabkan oleh adanya unsur acak.
